```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data=pd.read_csv('/content/wine_quality.csv')
data
```

```
data.info()
```

```
data.describe()
```

```
data.corr()['quality']
```

Heatmap correlation matrix

```
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(),annot=True,linewidth=1,center=0,cmap='coolwarm')
plt.show()
```

Histogram chart

```
plt.hist(data.quality,bins=6,alpha=0.5,histtype='bar',edgecolor='black')
plt.title('Quality of Wine')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

Bar chart for pH

```
plt.figure(figsize=(8,5))
sns.barplot(data['quality'],data['pH'],palette='GnBu_d')
plt.show()
```

From the barplot of Quality vs pH we can visualize that there is increse in pH with increse in quality

Box plot of quality vs pH

```
sns.boxplot(x='quality',y='pH',data=data,palette='GnBu_d')
plt.title('Boxplot of Quality and pH')
plt.show()
```

Boxplot of Quality and Alcohol

```
sns.boxplot(x='quality',y='alcohol',data=data,palette='GnBu_d')
plt.title('Boxplot of Quality and Alcohol')
plt.show()
```

The above Boxplot shows the increase in quality of wine with the increase in alcohol.

Boxplot of Quality and Residual sugar

```
sns.boxplot(x='quality',y='residual sugar',data=data,palette='GnBu_d')
plt.title('Boxplot of Quality and Residual sugar')
plt.show()
```

We can see there is not much effect of residual sugar on the quality of wine.

Boxplot of Quality and Density

```
sns.boxplot(x='quality',y='density',data=data,palette='GnBu_d')
plt.title('Boxplot of Quality and Density')
plt.show()
```

Greater the density of wine better will be the quality of wine.

Boxplot of Quality and Sulphates

```
sns.boxplot(x='quality',y='sulphates',data=data,palette='GnBu_d')
plt.title('Boxplot of Quality and Sulphates')
plt.show()
```

The above plot shows that the quality of wine increases with decrease in the amount of sulphates in the wine

Boxplot of Quality and Chlorides

```
sns.boxplot(x='quality',y='chlorides',data=data,palette='GnBu_d')
plt.title('Boxplot of Quality and Chlorides')
plt.show()
```

Boxplot of Quality and Ciric acid

```
sns.boxplot(x='quality',y='citric acid',data=data,palette='coolwarm')
plt.title('Boxplot of Quality and Citric acid')
plt.show()
```

The quality of the wine increses with increse in the amount of citric acid in the wine

Boxplot of Quality and Volatile acid

```
sns.boxplot(x='quality',y='volatile acidity',data=data,palette='coolwarm')
plt.title('Boxplot of Quality and Volatile acid')
plt.show()
```

From the above plot we can see that the quality of wine increses with the decrese in the amount of volatile acids 

Boxplot of Quality and Fixed acidity

```
sns.boxplot(x='quality',y='fixed acidity',data=data,palette='coolwarm')
plt.title('Boxplot of Quality and Fixed acidity')
plt.show()
```

there is no much effect of fixed acidity on wine
